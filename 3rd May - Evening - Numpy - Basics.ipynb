```python
#### Array Slicing (Picking Elements from the array)
```

```python
arr2 = np.array([[1,2,3,4,5,6],[4,5,6,2,4,5],[7,8,9,1,3,43]
                 ,[4,3,10,2,2,2],[44,33,22,2,2,1]])
arr2
```

```python
arr2[3]
```

```python
arr2[:]  ### all rows
```

```python
arr2[1:4]  ## index 1-3
```

```python
arr2[1:]  ## index 1 to last
```

```python
arr2[:4]  ## index 0 to 3
```

```python
arr2[:,:]  ## all rows and all cols (row,coln)
```

```python
arr2[1:4,1:4]  ## row - 1,2,3  coln - 1,2,3
```

```python
arr2[1:4,3]   ## coln 3 of row 1,2,3
```

```python
arr2[1:5]   ## row 1-4
```

```python
arr2[1:5:2]  ## 1,3
```

```python
arr2[1:5:3]  ## 1,4 
```
