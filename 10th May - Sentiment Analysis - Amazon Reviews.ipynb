```python
import nltk
import pandas as pd
```

```python
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()
```

```python
data = pd.read_csv("/content/amazon_reviews.tsv",sep='\t')
data.head()
```

```python
data['label'].value_counts()
```

```python
len(data)
```

```python
#data.iloc[0,1]
data.iloc[0]['review']
sia.polarity_scores(data.iloc[0]['review'])
```

```python
### Calculate the Polarity Score for each Review
pol = []
for i in range(len(data)):
  pol.append(sia.polarity_scores(data.iloc[i,1]))
```

```python
data['Polarity Scores'] = pol
data.head()
```

```python
s = data['Polarity Scores'].iloc[0]
s['compound']
```

```python
comp = []
for i in range(len(data)):
  s = data['Polarity Scores'].iloc[i]
  comp.append(s['compound'])
```

```python
data['Compound'] = comp
data.head()
```

```python
sent = []
for i in range(len(data)):
  s = data['Polarity Scores'].iloc[i]
  if s['compound'] > 0:
    sent.append('pos')
  else:
    sent.append('neg')
```

```python
data['Predicted Sentiment'] = sent
data.head(20)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(data['label'],data['Predicted Sentiment'])
cm
```

```python
from sklearn.metrics import accuracy_score
a = accuracy_score(data['label'],data['Predicted Sentiment'])
a
```
