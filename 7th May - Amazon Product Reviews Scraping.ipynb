```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.amazon.in/Samsung-Galaxy-Ocean-128GB-Storage/product-reviews/B07HGGYWL6/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews"
code = requests.get(url)
code
```

```python
soup = BeautifulSoup(code.content,'html.parser')
print(soup.prettify())
```

```python
names = soup.select('span.a-profile-name')[2:]
names
```

```python
len(names)
```

```python
stars = soup.select('i.review-rating span.a-icon-alt')[2:]
stars
```

```python
len(stars)
```

```python
titles = soup.select('a.review-title span')
titles
```

```python
len(titles)
```

```python
dates = soup.select('span.review-date')[2:]
dates
```

```python
len(dates)
```

```python
reviews = soup.select('span.review-text-content span')
reviews
```

```python
len(reviews)
```

```python
names[0].get_text()
```

```python
cust_name = []
rev_date = []
rev_title = []
rev_stars = []
rev_content = []
for i in range(len(names)):
  cust_name.append(names[i].get_text())
  rev_date.append(dates[i].get_text().lstrip("Reviewed in India on "))
  rev_title.append(titles[i].get_text())
  rev_stars.append(stars[i].get_text().replace(" out of 5 stars",""))
  rev_content.append(reviews[i].get_text().strip("\n  "))
```

```python
rev_content
```

```python
import pandas as pd
```

```python
df = pd.DataFrame()
df['Date'] = rev_date
df['Customer Name'] = cust_name
df['Ratings'] = rev_stars
df['Review Title'] = rev_title
df['Reviews'] = rev_content
df
```

```python
df.to_csv("amazon.csv")
```
