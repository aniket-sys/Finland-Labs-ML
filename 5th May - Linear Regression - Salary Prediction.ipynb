```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/content/Salary_Data.csv")
data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.head()
```

```python
len(data)
```

```python
### Shape of Feature should always be (num.of samples,no.of features) -- 2D Array
### Shape of Target should always be either (no.of samples,) or (no.of samples,no.of targets) -- 1d or 2D
```

```python
### Shape of feature - (30,1)
### Shape of target - (30,) or (30,1)
```

```python
### Divide the data into features(ip) and target(op)
x = data['YearsExperience'].values
x
```

```python
x = x.reshape(len(data),1)
x.shape
```

```python
y = data['Salary'].values
y
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,train_size=0.8,random_state=200)
```

```python
xtest
```

```python
#### Build the model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
### Perform Training 
model.fit(xtrain,ytrain)
```

```python
### Prediction
ypred = model.predict(xtest)
ypred
```

```python
xtest
```

```python
ytest
```

```python
### R2 Score
from sklearn.metrics import r2_score
r2 = r2_score(ytest,ypred)
r2*100
```

```python
#### Line of Regression (Training Samples)
plt.scatter(xtrain,ytrain,color="red",label="Actual Points")
plt.scatter(xtrain,model.predict(xtrain),color="blue",label="Predicted Points")
plt.plot(xtrain,model.predict(xtrain),color="orange",label="Line of Regression")
plt.legend()
plt.show()
```

```python
#### Line of Regression (Testing Samples)
plt.scatter(xtest,ytest,color="red",label="Actual Points")
plt.scatter(xtest,model.predict(xtest),color="blue",label="Predicted Points")
plt.plot(xtest,model.predict(xtest),color="orange",label="Line of Regression")
plt.legend()
plt.show()
```

```python
m = model.coef_   ## slope
c = model.intercept_   ## y intercept
```

```python
print(m,c)
```

```python
m * 10.1 + c  ## salary for 10.1
```

```python
score = []
for i in range(501):
  xtrain_1,xtest_1,ytrain_1,ytest_1 = train_test_split(x,y,train_size=0.8,random_state=i)
  model_1 = LinearRegression()
  model_1.fit(xtrain_1,ytrain_1)
  ypred_1 = model_1.predict(xtest_1)
  r2_new = r2_score(ytest_1,ypred_1)
  score.append(r2_new)
```

```python
score
```

```python
np.max(score)
```
