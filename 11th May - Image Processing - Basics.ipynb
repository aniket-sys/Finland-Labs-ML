#### Draw Different Shapes on Images

```python
### Create a blank white image (512 x 512 x 3)
img = np.ones((512,512,3),np.uint8)*255
plt.imshow(img)
plt.show()
```

```python
### Draw a line
img = np.ones((512,512,3),np.uint8)*255
## (img,pt1,pt2,colour,thickness)
cv2.line(img,(10,200),(300,30),(255,0,0),5)
plt.imshow(img)
plt.show()
```

```python
### Draw a rectangle/square
img = np.ones((512,512,3),np.uint8)*255
## (img,pt1,pt2,colour,thickness)
cv2.rectangle(img,(10,200),(300,30),(255,0,0),5)
plt.imshow(img)
plt.show()
```

```python
### Draw a rectangle/square
img = np.ones((512,512,3),np.uint8)*255
## (img,pt1,pt2,colour,thickness)
cv2.rectangle(img,(10,200),(300,30),(50,80,100),-1)
plt.imshow(img)
plt.show()
```

```python
### Draw a Circle
img = np.ones((512,512,3),np.uint8)*255
## (img,center,radius,colour,thickness)
cv2.circle(img,(250,250),200,(50,80,100),3)
cv2.circle(img,(250,250),10,(50,80,100),-1)
plt.imshow(img)
plt.show()
```

```python
### Draw an Ellipse
img = np.ones((512,512,3),np.uint8)*255
## (img, center, axes, angle, startAngle, endAngle, color, thickness)
cv2.ellipse(img,(250,250),(200,50),0,0,360,(50,80,100),3)
plt.imshow(img)
plt.show()
```

```python
### Draw an Ellipse
img = np.ones((512,512,3),np.uint8)*255
## (img, center, axes, angle, startAngle, endAngle, color, thickness)
cv2.ellipse(img,(250,250),(200,50),0,0,360,(50,80,100),-1)
plt.imshow(img)
plt.show()
```

```python
### Draw an Polygon
img = np.ones((512,512,3),np.uint8)*255
## (img, [points],isclosed, color, thickness)
points = np.array([[10,200],[80,80],[400,200]])
cv2.polylines(img,[points],True,(50,80,100),3)
plt.imshow(img)
plt.show()
```

```python
### Draw an Polygon
img = np.ones((512,512,3),np.uint8)*255
## (img, [points],isclosed, color, thickness)
points = np.array([[10,200],[80,80],[400,200]])
cv2.polylines(img,[points],False,(50,80,100),3)
plt.imshow(img)
plt.show()
```

```python
### Write Text on Image
img = np.ones((512,512,3),np.uint8)*255
## (img,text,position,fontstyle,fontsize, color, thickness)
cv2.putText(img,"Umang",(50,200),cv2.FONT_HERSHEY_SCRIPT_COMPLEX,2.5,(50,80,100),5)
plt.imshow(img)
plt.show()
```
