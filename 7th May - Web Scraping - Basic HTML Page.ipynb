```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "http://finlandlabs.com/demo.html"
code = requests.get(url)
code
```

```python
code.content
```

```python
soup = BeautifulSoup(code.content,'html.parser')
soup
```

```python
print(soup.prettify())
```

```python
data = soup.find_all('p')
data
```

```python
data = soup.find_all('a')
data
```

```python
data = soup.find_all('p',class_='inner-text')
data
```

```python
data = soup.find_all('p',class_='inner-text first-item')
data
```

```python
data = soup.find_all('p',id = 'first')
data
```
