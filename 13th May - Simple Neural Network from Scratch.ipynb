```python
import numpy as np
```

```python
x = np.array([[1,0,1],[0,0,1],[0,1,0],[1,1,0]])
y = np.array([[0],[0],[1],[1]])
```

```python
## Assigning the Weights
np.random.seed(10)
weights = np.random.random((3,1))
weights
```

```python
sum = np.dot(x,weights) + 0.02 ## 0.02-bias
sum
```

```python
## Activation (sigmoid)
def sig(x):
  return 1/(1+np.exp(-x))
```

```python
ypred = sig(sum)
ypred
```

```python
ypred.round()
```

```python
sum = np.dot(x,weights) + 0.02
ypred = sig(sum)
error = y - ypred
error
```

```python
def gradient(a,x):
  return a * x * (1-x)
```

```python
sum = np.dot(x,weights) + 0.02
ypred = sig(sum)
error = y - ypred
change_in_weights = gradient(error,ypred)
weights = weights + np.dot(x.T,change_in_weights)
error
```

```python
for i in range(500):
  sum = np.dot(x,weights) + 0.02
  ypred = sig(sum)
  error = y - ypred
  change_in_weights = gradient(error,ypred)
  weights = weights + np.dot(x.T,change_in_weights)
```
