```python
### Drop unwanted features
data = data.drop(['PassengerId','Name','Ticket','Fare','Cabin'],axis=1)
```

```python
data.head()
```

```python
### Missing Values
data.isnull().sum()
```

```python
mean_age = data['Age'].mean()
mean_age = round(mean_age,2)
mean_age
```

```python
data['Age'] = data['Age'].fillna(mean_age)
```

```python
data.isnull().sum()
```

```python
data = data.dropna()
```

```python
data.isnull().sum()
```

```python
len(data)
```

```python
data.head()
```

```python
### Label Encoding
data['Sex'] = data['Sex'].map({"male":0,"female":1})
data['Embarked'] = data['Embarked'].map({"S":0,"C":1,"Q":2})
```

```python
data.head()
```

```python
#scaling
#from sklearn.preprocessing import StandardScaler
#ss = StandardScaler()
#data['Age'] = ss.fit_transform(data[['Age']])
```

```python
data.head()
```

```python
#### Divide data into features and target
x = data.iloc[:,1:].values
y = data.iloc[:,0].values
```

```python
### Split the data into training and testing
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,train_size=0.85,random_state=3) 
```

```python
pd.DataFrame(y).value_counts()
```

```python
pd.DataFrame(ytrain).value_counts()
```

```python
### Build the model
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=9,p=1)  ### p=2(euclidean) p=1(manhattan)
```

```python
### Train the model
model.fit(xtrain,ytrain)
```

```python
ypred = model.predict(xtest)
ypred
```

```python
ytest
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(ytest,ypred)
cm
```

```python
from sklearn.metrics import accuracy_score
a = accuracy_score(ytest,ypred)
a
```

```python
len(ytrain)
```

```python
accuracy = []
for k in range(3,200):
  model1 = KNeighborsClassifier(n_neighbors=k,p=2)
  model1.fit(xtrain,ytrain)
  ypred1 = model1.predict(xtest)
  a1 = accuracy_score(ytest,ypred1)
  accuracy.append(a1)
```

```python
accuracy
```

```python
np.max(accuracy)
```

```python
np.argmax(accuracy)
```
