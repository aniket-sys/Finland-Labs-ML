```python
import matplotlib.pyplot as plt
```

```python
### Line Graph
### Bar Graph
### Scatter Plot
## Pie Chart
## Subplot
```

```python
### Line Graph
x = [3,4,5,6,7,1,2,8]
y = [33,21,11,34,22,45,66,88]
plt.figure(figsize=(8,4))
plt.title("Line Graph",fontsize=15,color='blue')
plt.xlabel("X Axis",fontsize=13,color='blue')
plt.ylabel("Y Axis",fontsize=13,color='blue')
plt.plot(x,y,color="red",lw=3,linestyle="dotted",label="Line Graph")  ## lw=linewidth  
## style  - solid,dashed,dotted
plt.legend(loc='best')
plt.show()
```

```python
### Verical Bar Graph
x = [3,4,5,6,7,1,2,8]
y = [33,21,11,34,22,45,66,88]
plt.figure(figsize=(8,4))
plt.title("Line Graph",fontsize=15,color='blue')
plt.xlabel("X Axis",fontsize=13,color='blue')
plt.ylabel("Y Axis",fontsize=13,color='blue')
plt.bar(x,y,color=["red","orange"],label="Bar Graph",width=0.5)  
plt.legend(loc='best')
plt.show()
```

```python
### Vertical Bar Graph
x = [3,4,5,6,7,1,2,8]
y = [33,21,11,34,22,45,66,88]
plt.figure(figsize=(8,4))
plt.title("Line Graph",fontsize=15,color='blue')
plt.xlabel("Y Axis",fontsize=13,color='blue')
plt.ylabel("X Axis",fontsize=13,color='blue')
plt.barh(x,y,color=["red","orange"],label="Bar Graph",height=0.5)  
plt.legend(loc='best')
plt.show()
```

```python
### Scatter
x = [3,4,5,6,7,1,2,8]
y = [33,21,11,34,22,45,66,88]
plt.figure(figsize=(8,4))
plt.title("Line Graph",fontsize=15,color='blue')
plt.xlabel("X Axis",fontsize=13,color='blue')
plt.ylabel("Y Axis",fontsize=13,color='blue')
plt.scatter(x,y,color="orange",label="Scatter Graph",s=200,marker="*")  ## s-size  
## marker -- o,*,d,s,<,>,^,v,x,+
plt.legend(loc='best')
plt.show()
```

```python
### Pie Chart
slices = [10,30,50,80,100]
names = ["C","C++","Java","Javascript","Python"]
cols = ["red","orange","blue","purple","green"]
plt.figure(figsize=(6,6))  ## width,height in inches
plt.pie(slices,colors = cols,labels=names,autopct="%0.2f%%",
        textprops={"color":"maroon","fontsize":12},explode=(0,0.5,0,0,0))
plt.show()
```

```python
#### Subplot  -- 3 rows and 3 cols

x = [3,4,5,6,7,1,2,8]
y = [33,21,11,34,22,45,66,88]
plt.figure(figsize=(12,12))  ## width,height in inches
plt.subplot(3,3,1)   ### rows,cols,position
plt.title("Graph 1")
plt.plot(x,y,color="red",lw=3,linestyle="dotted",label="Line Graph")

plt.subplot(3,3,3)
plt.bar(x,y,color=["red","orange"],label="Bar Graph",width=0.5)

plt.subplot(3,3,5)
plt.barh(x,y,color=["red","orange"],label="Bar Graph",height=0.5) 

plt.subplot(3,3,7)
plt.scatter(x,y,color="orange",label="Scatter Graph",s=200,marker="*")

plt.subplot(3,3,9)
slices = [10,30,50,80,100]
names = ["C","C++","Java","Javascript","Python"]
cols = ["red","orange","blue","purple","green"]
plt.pie(slices,colors = cols,labels=names,autopct="%0.2f%%",
        textprops={"color":"maroon","fontsize":12},explode=(0,0.5,0,0,0))
plt.savefig("mygraph.png")
plt.show()
```
